### Step 4. How many products cost more than $10.00?

```python
# clean the item_price column and transform it in a float
prices = [float(value[1 : -1]) for value in chipo.item_price]  #將夾雜的錢字符號$去除, 只取數字
print(type(prices))  #變成一個list, 但格式為字串不是數值喔

#pp = pprint.PrettyPrinter(indent=4)
pp = pprint.PrettyPrinter(width=100, compact=True)
pp.pprint(prices)
```

```python
# reassign the column with the cleaned prices
chipo.item_price = prices #將清理掉錢符號的資料覆寫回去該欄  但格式為字串不是數值喔
```

```python
chipo.describe()
```

```python
#將整個檔中, 數量相同的商品名稱(去重複) , 不管訂單號碼   ????  寫到 chipo_filtered

# delete the duplicates in item_name and quantity
chipo_filtered = chipo.drop_duplicates(['item_name','quantity','choice_description'])

print(len(chipo_filtered))  #筆數只剩 1949
print(chipo_filtered)  #因為螢幕寬度 印出來很醜
```

```python
#刪之前每張訂單有幾筆, count 完變成serier; 要加 .to_frame() 以利後續merge
chipo_count_perorder_b4 = chipo.groupby('order_id').quantity.count().to_frame()


#刪之後每張訂單有幾筆
chipo_count_perorder_aft = chipo_filtered.groupby('order_id').quantity.count().to_frame()

#有差異的訂單是哪些
chipo_count_perorder_diff =chipo_count_perorder_b4.merge(chipo_count_perorder_aft, left_on='order_id', right_on='order_id')
#print(chipo_count_perorder_diff)

#merge的時候,若有相同欄名會自動加底線x,y

#print(chipo_count_perorder_diff)
#filter = chipo_count_perorder_diff["quantity_x"] != chipo_count_perorder_diff["quantity_y"]
#chipo_count_perorder_diff.where(filter, inplace = True)

chipo_count_perorder_diff['variance'] = chipo_count_perorder_diff['quantity_x'] - chipo_count_perorder_diff['quantity_y']
#chipo_count_perorder_diff[chipo_count_perorder_diff['variance'] > 0]
```

```python
# select only the products with quantity equals to 1
chipo_one_prod = chipo_filtered[chipo_filtered.quantity == 1]
chipo_one_prod

# chipo_one_prod[chipo_one_prod['item_price']>10].item_name.nunique()
# chipo_one_prod[chipo_one_prod['item_price']>10]
```
